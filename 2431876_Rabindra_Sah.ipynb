## Task-3 - Outlier Detection:

## 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
## 2. Define outliers using the 1.5  IQR rule.
## 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
## 4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
def find_outliers(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bond = Q3 + 1.5 * IQR
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bond)]
  return outliers, lower_bound, upper_bond

  # Identify outliers for 'Score' and 'GDP per Capita'
score_outliers, score_lower, score_upper = find_outliers(df, 'score')
gdp_outliers, gdp_lower, gdp_upper = find_outliers(df, 'Log GDP per capita')

# Combine outliers
combined_outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()
print("Outlier countries based on Score and GDP per Capita:\n", combined_outliers[['Country name', 'score', 'Log GDP per capita']])

# Step 3: Scatter Plot highlighting outliers
plt.figure(figsize=(10, 6))

# Scatter plot for all countries
sns.scatterplot(data=df, x='Log GDP per capita', y='score', color='skyblue', label='South Asia')

# Highlight outliers in red
sns.scatterplot(data=combined_outliers, x='Log GDP per capita', y='score', color='red', s=100, label='Outliers')

plt.title('Scatter Plot of GDP per Capita vs Score (Outliers Highlighted)')
plt.xlabel('GDP per Capita (Log)')
plt.ylabel('Happiness Score')
plt.legend()
plt.show()

# Step 4: Discussion of Outliers
print("Lower and Upper Bound for Score: ", score_lower, score_upper)
print("Lower and Upper Bound for GDP per Capita: ", gdp_lower, gdp_upper)

# Display characteristics of outliers
for _, row in combined_outliers.iterrows():
    print(f"Country: {row['Country name']}, Score: {row['score']}, GDP per Capita: {row['Log GDP per capita']}")
```

```python
"""
This code finds the correlation between happiness Score and metrics such as Freedom to make life choices and Generosity.
Steps:
1. Calculates the Pearson correlation coefficients for each metric with 'score'.
2. Visualizes the relationships using regression plots.
3. Identifies the metric with the stronger correlation to 'score'.
4. Discusses the findings with numerical correlation values.
"""
# Defines the matrics
metrics = ['Freedom to make life choices', 'Generosity']
for metric in metrics:
  coorelation = df['score'].corr(df[metric])
  print(f"Correlation between 'score' and '{metric}': {coorelation}")
#Set up the plotting area for two subplots
plt.figure(figsize=(12, 5))
# plot the correlation between 'score' and 'Freedom to make life choices'
plt.subplot(1, 2, 1)
sns.regplot(data=df, x='Freedom to make life choices', y='score', color='green')
plt.title('coorelation between score and Freedom to make life choices')
plt.xlabel('Freedom to make life choices')
plt.ylabel('score')
## plot the correlation between 'score' and 'Generosity'
plt.subplot(1, 2, 2)
sns.regplot(data=df, x='Generosity', y='score', color='lightcoral')
plt.title('Score vs Generosity')
plt.xlabel('Generosity')
plt.ylabel('Score')

plt.tight_layout()
plt.show()

#Identify the strongest and weakest correlation
freedom_corr = df['score'].corr(df['Freedom to make life choices'])
generosity_corr = df['score'].corr(df['Generosity'])

if abs(freedom_corr) > abs(generosity_corr):
    print("The strongest relationship is between Score and Freedom to Make Life Choices.")
else:
    print("The weakest relationship is between Score and Generosity.")

print("Discussion:")
print(f"1. Freedom to Make Life Choices has a correlation of {freedom_corr:.3f} with Score.")
print(f"2. Generosity has a correlation of {generosity_corr:.3f} with Score.")
```

```python
def identify_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bond = Q1 - 1.5 * IQR
    upper_bond = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bond) | (data[column] > upper_bond)]
    return outliers

#Finding outliers for South Asia
south_asia_outliers_for_score = identify_outliers(df, 'score')
south_asia_outliers_for_GDP = identify_outliers(df, 'Log GDP per capita')

#Finding outliers for Middle East
middle_east_outliers_for_score = identify_outliers(filtered_middle_east_countries, 'score')
middle_east_outliers_for_GDP = identify_outliers(filtered_middle_east_countries, 'Log GDP per capita')

#combining outliers
combined_outliers = pd.concat([
    south_asia_outliers_for_score.assign(Region='South Asia', Metric='score'),
    south_asia_outliers_for_GDP.assign(Region='South Asia', Metric='Log GDP per capita'),
    middle_east_outliers_for_score.assign(Region='Middle East', Metric='score'),
    middle_east_outliers_for_GDP.assign(Region='Middle East', Metric='Log GDP per capita')
])

print("Outliers detected based on Score and GDP per Capita:")
print(combined_outliers[['Country name', 'Region', 'Metric', 'score', 'Log GDP per capita']])

plt.subplot(1,2,1)
plt.scatter(x='Log GDP per capita', y='score', data=df, color='blue', label='South Asia')
plt.scatter(x='Log GDP per capita', y='score', data=south_asia_outliers_for_score, color='red', label='Score outliers', s=100)
plt.scatter(x='Log GDP per capita', y='score', data=south_asia_outliers_for_GDP, color='green', label='GDP per capita outliers', s=100)
plt.title("South Asia: outliers for Score and GDP per Capita")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()

#Visualizing the outliers by scatter plot
plt.subplot(1,2,2)
plt.scatter(x='Log GDP per capita', y='score', data=filtered_middle_east_countries, color='blue', label='South Asia')
plt.scatter(x='Log GDP per capita', y='score', data=middle_east_outliers_for_score, color='red', label='Score outliers', s=100)
plt.scatter(x='Log GDP per capita', y='score', data=middle_east_outliers_for_GDP, color='green', label='GDP per capita outliers', s=100)
plt.title("Middle East: outliers for Score and GDP per Capita")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()

plt.show()
```
